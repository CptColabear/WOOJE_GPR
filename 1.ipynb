```python
import os
import platform
import shutil
import subprocess
import sys
from datetime import datetime

def run(cmd):
    try:
        return subprocess.check_output(
            cmd, stderr=subprocess.STDOUT, text=True
        ).strip()
    except Exception:
        return None

def bytes_to_gb(b):
    return b / (1024 ** 3)

print("====================================")
print(" System Spec Snapshot (Ubuntu Server)")
print("====================================")
print("Time:", datetime.now().isoformat(timespec="seconds"))
print("Python:", sys.version.replace("\n", " "))
print("OS:", platform.platform())
print("Kernel:", platform.release())
print("Architecture:", platform.machine())

# ---------------- CPU ----------------
print("\n[CPU]")
print("Logical cores:", os.cpu_count())

cpu_model = run(["bash", "-lc", "lscpu | grep 'Model name'"])
if cpu_model:
    print(cpu_model.strip())

# ---------------- RAM ----------------
print("\n[Memory]")
ram_gb = None
if os.path.exists("/proc/meminfo"):
    with open("/proc/meminfo") as f:
        for line in f:
            if line.startswith("MemTotal:"):
                kb = int(line.split()[1])
                ram_gb = kb * 1024 / (1024 ** 3)
                break

print(f"Total RAM: {ram_gb:.2f} GB" if ram_gb else "RAM: N/A")

# ---------------- Disk ----------------
print("\n[Disk]")
total, used, free = shutil.disk_usage("/")
print(f"Root(/) total: {bytes_to_gb(total):.2f} GB")
print(f"Root(/) free : {bytes_to_gb(free):.2f} GB")

# SSD / HDD 여부 (best-effort)
print("\n[Disk Type]")
rota = run(["bash", "-lc", "lsblk -d -o name,rota"])
if rota:
    print("name rota (0=SSD/NVMe, 1=HDD)")
    print(rota)

# ---------------- GPU ----------------
print("\n[GPU]")
nvidia = run([
    "nvidia-smi",
    "--query-gpu=name,memory.total,driver_version",
    "--format=csv,noheader"
])
if nvidia:
    print("NVIDIA GPU detected:")
    print(nvidia)
else:
    print("No NVIDIA GPU (or nvidia-smi not installed)")

# ---------------- MongoDB ----------------
print("\n[MongoDB]")
mongod_version = run(["mongod", "--version"])
if mongod_version:
    print(mongod_version)
else:
    print("mongod not found (MongoDB not installed or PATH issue)")

print("\nDone.")
```
