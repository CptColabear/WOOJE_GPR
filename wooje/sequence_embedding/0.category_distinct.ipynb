```python
from pymongo import MongoClient

client = MongoClient("mongodb://10.255.68.40:27017")
db = client["ejoow"]
col = db["0. US_category_frequency"]

# 전체 distinct
cats = col.distinct("VenueCategoryname")
print("Num distinct categories:", len(cats))
print("Sample:", cats[:30])
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

def tok_info(text: str):
    toks = tokenizer.tokenize(text)
    return toks

multi = []
subword_split = []  # '##' 포함
single = []

for c in cats:
    toks = tok_info(c)
    if len(toks) == 1:
        single.append((c, toks))
    else:
        multi.append((c, toks))
        if any(t.startswith("##") for t in toks):
            subword_split.append((c, toks))

print("Distinct categories:", len(cats))
print("Single-token:", len(single))
print("Multi-token:", len(multi))
print("Has ## subword split:", len(subword_split))

print("\nExamples of multi-token (first 20):")
for c, toks in multi[:20]:
    print(f"- {c} -> {toks}")

print("\nExamples of ## subword splits (first 20):")
for c, toks in subword_split[:20]:
    print(f"- {c} -> {toks}")
```

```python
with open("venue_categories.txt", "w", encoding="utf-8") as f:
    for c in sorted(cats):
        f.write(c + "\n")

print("Saved to venue_categories.txt")
```
