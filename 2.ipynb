```python
from pymongo import MongoClient
from pprint import pprint

MONGO_HOST = '10.255.68.40'
MONGO_PORT = 27017
DB_NAME = 'ejoow'
COLLECTION_NAME = '0. US_group_checkin_labeled'
GROUP_FIELD = 'Group_ID'

client = MongoClient(f'mongodb://{MONGO_HOST}:{MONGO_PORT}/')
db = client[DB_NAME]
col = db[COLLECTION_NAME]

print(f'Connected to {DB_NAME}.{COLLECTION_NAME}')
```

```python
# 샘플 문서 확인
sample = col.find_one()
pprint(sample)
```

```python
# 전체 문서 수
total_docs = col.count_documents({})
print('Total documents:', total_docs)
```

```python
# Group_ID 기준 고유 그룹 수 계산
unique_group_count = len(col.distinct(GROUP_FIELD))
print(f'Unique group count based on {GROUP_FIELD}:', unique_group_count)
```

```python
# aggregation으로 한 번 더 검증
pipeline = [
    {'$group': {'_id': f'${GROUP_FIELD}'}},
    {'$count': 'num_groups'}
]

result = list(col.aggregate(pipeline))
agg_count = result[0]['num_groups'] if result else 0
print('Unique group count by aggregation:', agg_count)
```

```python
from pymongo import MongoClient
from pprint import pprint

MONGO_HOST = '10.255.68.40'
MONGO_PORT = 27017
DB_NAME = 'ejoow'
COLLECTION_NAME = '2. US_checkin_v2'
USER_FIELD = 'user_id'

client = MongoClient(f'mongodb://{MONGO_HOST}:{MONGO_PORT}/')
db = client[DB_NAME]
col = db[COLLECTION_NAME]

print(f'Connected to {DB_NAME}.{COLLECTION_NAME}')
```

```python
sample = col.find_one()
pprint(sample)

total_docs = col.count_documents({})
print('Total check-in documents:', total_docs)

unique_user_count = len(col.distinct(USER_FIELD))
print(f'Unique user count based on {USER_FIELD}:', unique_user_count)
```

```python
pipeline = [
    {'$group': {'_id': f'${USER_FIELD}'}},
    {'$count': 'num_users'}
]

result = list(col.aggregate(pipeline, allowDiskUse=True))
agg_count = result[0]['num_users'] if result else 0
print('Unique user count by aggregation:', agg_count)
```
